```python
import requests
import matplotlib.pyplot as plt
import time
```

```python
timestamps = []
latitudes = []
longitudes = []

points = 100
```

```python
for i in range(points):
    response = requests.get("https://api.wheretheiss.at/v1/satellites/25544")
    if response.status_code == 200:
        data = response.json()
        timestamp = data["timestamp"]
        latitude = data["latitude"]
        longitude = data["longitude"]

        timestamps.append(timestamp)
        latitudes.append(latitude)
        longitudes.append(longitude)

        if i == 0:
            start_timestamp = timestamp
            
        elapsed_seconds = timestamp - start_timestamp

        print("Point " + str(i + 1) + "/" + str(points) + " - Time Elapsed: " + str(elapsed_seconds) + " seconds, " + "Latitude: " + str(round(latitude, 2)) + ", " + "Longitude: " + str(round(longitude, 2)))
    else:
        print("Error fetching data.")

    if i < points - 1:
        time.sleep(10)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(longitudes, latitudes, marker='o', linestyle='-', markersize=4)
plt.title("International Space Station (ISS) Path")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```
